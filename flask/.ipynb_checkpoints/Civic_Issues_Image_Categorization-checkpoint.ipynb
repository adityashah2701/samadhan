```python
# Civic Categories Multi-Class Classification using Transfer Learning
# 7 Classes: garbage bin, road, sidewalk path, drainage system, parking areas, street lights, park

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
tf.random.set_seed(42)
np.random.seed(42)
```

```python
# Configuration
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
EPOCHS = 20
NUM_CLASSES = 7

# Class names
CLASS_NAMES = [
    'garbage_bin',
    'road', 
    'sidewalk_path',
    'drainage_system',
    'parking_areas',
    'street_lights',
    'park'
]

print("🏗️ Civic Categories Classification Model")
print("=" * 50)
print(f"Classes: {CLASS_NAMES}")
print(f"Image Size: {IMG_SIZE}")
print(f"Batch Size: {BATCH_SIZE}")
```

```python
DATA_DIR = "civic_categories"  # Update this to your dataset path
TRAIN_DIR = os.path.join(DATA_DIR, "train")
VAL_DIR = os.path.join(DATA_DIR, "validation")
```

```python
# Data Augmentation and Preprocessing
def create_data_generators():
    """Create data generators with augmentation for training and validation"""
    
    # Training data generator with augmentation
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        horizontal_flip=True,
        zoom_range=0.2,
        shear_range=0.2,
        fill_mode='nearest',
        brightness_range=[0.8, 1.2]
    )
    
    # Validation data generator (only rescaling)
    val_datagen = ImageDataGenerator(rescale=1./255)
    
    # Create generators
    train_generator = train_datagen.flow_from_directory(
        TRAIN_DIR,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        classes=CLASS_NAMES,
        shuffle=True
    )
    
    validation_generator = val_datagen.flow_from_directory(
        VAL_DIR,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        classes=CLASS_NAMES,
        shuffle=False
    )
    
    return train_generator, validation_generator
```

```python
# Model Architecture using Transfer Learning
def create_model():
    """Create the model using EfficientNetB0 as base model"""
    
    # Load pre-trained EfficientNetB0 model
    base_model = EfficientNetB0(
        weights='imagenet',
        include_top=False,
        input_shape=(*IMG_SIZE, 3)
    )
    
    # Freeze the base model
    base_model.trainable = False
    
    # Add custom classification head
    model = keras.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        layers.Dense(512, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(NUM_CLASSES, activation='softmax', name='predictions')
    ])
    
    return model

# Compile model
def compile_model(model):
    """Compile the model with appropriate optimizer and loss function"""
    
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss='categorical_crossentropy',
        metrics=['accuracy', 'top_2_accuracy']
    )
    
    return model
```

```python
# Training callbacks
def create_callbacks():
    """Create callbacks for training"""
    
    callbacks = [
        keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=5,
            restore_best_weights=True,
            verbose=1
        ),
        keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=3,
            min_lr=1e-7,
            verbose=1
        ),
        keras.callbacks.ModelCheckpoint(
            'civic_classifier_best.h5',
            monitor='val_accuracy',
            save_best_only=True,
            verbose=1
        )
    ]
    
    return callbacks
```

```python
# Visualization functions
def plot_training_history(history):
    """Plot training and validation metrics"""
    
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    fig.suptitle('Training History', fontsize=16)
    
    # Accuracy
    axes[0,0].plot(history.history['accuracy'], label='Training Accuracy')
    axes[0,0].plot(history.history['val_accuracy'], label='Validation Accuracy')
    axes[0,0].set_title('Model Accuracy')
    axes[0,0].set_xlabel('Epoch')
    axes[0,0].set_ylabel('Accuracy')
    axes[0,0].legend()
    axes[0,0].grid(True)
    
    # Loss
    axes[0,1].plot(history.history['loss'], label='Training Loss')
    axes[0,1].plot(history.history['val_loss'], label='Validation Loss')
    axes[0,1].set_title('Model Loss')
    axes[0,1].set_xlabel('Epoch')
    axes[0,1].set_ylabel('Loss')
    axes[0,1].legend()
    axes[0,1].grid(True)
    
    # Top-2 Accuracy
    axes[1,0].plot(history.history['top_2_accuracy'], label='Training Top-2 Acc')
    axes[1,0].plot(history.history['val_top_2_accuracy'], label='Validation Top-2 Acc')
    axes[1,0].set_title('Top-2 Accuracy')
    axes[1,0].set_xlabel('Epoch')
    axes[1,0].set_ylabel('Top-2 Accuracy')
    axes[1,0].legend()
    axes[1,0].grid(True)
    
    # Learning Rate
    if 'lr' in history.history:
        axes[1,1].plot(history.history['lr'], label='Learning Rate')
        axes[1,1].set_title('Learning Rate')
        axes[1,1].set_xlabel('Epoch')
        axes[1,1].set_ylabel('Learning Rate')
        axes[1,1].set_yscale('log')
        axes[1,1].legend()
        axes[1,1].grid(True)
    
    plt.tight_layout()
    plt.show()

def plot_confusion_matrix(y_true, y_pred, class_names):
    """Plot confusion matrix"""
    
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
# Prediction functions
def preprocess_image(image_path):
    """Preprocess a single image for prediction"""
    
    img = Image.open(image_path).convert('RGB')
    img = img.resize(IMG_SIZE)
    img_array = np.array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    
    return img_array

def predict_single_image(model, image_path, show_image=True):
    """Predict class for a single image"""
    
    # Preprocess image
    img_array = preprocess_image(image_path)
    
    # Make prediction
    predictions = model.predict(img_array, verbose=0)
    predicted_class_idx = np.argmax(predictions[0])
    predicted_class = CLASS_NAMES[predicted_class_idx]
    confidence = predictions[0][predicted_class_idx]
    
    # Get top 3 predictions
    top_3_idx = np.argsort(predictions[0])[-3:][::-1]
    top_3_predictions = [(CLASS_NAMES[idx], predictions[0][idx]) for idx in top_3_idx]
    
    if show_image:
        # Display image with prediction
        img = Image.open(image_path)
        plt.figure(figsize=(12, 6))
        
        plt.subplot(1, 2, 1)
        plt.imshow(img)
        plt.title(f'Input Image\nPredicted: {predicted_class}\nConfidence: {confidence:.2%}')
        plt.axis('off')
        
        plt.subplot(1, 2, 2)
        classes = [pred[0] for pred in top_3_predictions]
        confidences = [pred[1] for pred in top_3_predictions]
        
        plt.barh(classes, confidences)
        plt.xlabel('Confidence')
        plt.title('Top 3 Predictions')
        plt.xlim(0, 1)
        
        # Add percentage labels
        for i, v in enumerate(confidences):
            plt.text(v + 0.01, i, f'{v:.1%}', va='center')
        
        plt.tight_layout()
        plt.show()
    
    return predicted_class, confidence, top_3_predictions
```

```python
# Fine-tuning function
def fine_tune_model(model, train_generator, validation_generator, epochs=10):
    """Fine-tune the model by unfreezing some layers"""
    
    print("🔧 Starting Fine-tuning...")
    
    # Unfreeze the top layers of the base model
    base_model = model.layers[0]
    base_model.trainable = True
    
    # Freeze all layers except the last 20
    for layer in base_model.layers[:-20]:
        layer.trainable = False
    
    # Compile with a lower learning rate
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.0001/10),
        loss='categorical_crossentropy',
        metrics=['accuracy', 'top_2_accuracy']
    )
    
    # Create callbacks for fine-tuning
    fine_tune_callbacks = [
        keras.callbacks.EarlyStopping(
            monitor='val_accuracy',
            patience=5,
            restore_best_weights=True
        ),
        keras.callbacks.ModelCheckpoint(
            'civic_classifier_fine_tuned.h5',
            monitor='val_accuracy',
            save_best_only=True,
            verbose=1
        )
    ]
    
    # Fine-tune the model
    history_fine = model.fit(
        train_generator,
        epochs=epochs,
        validation_data=validation_generator,
        callbacks=fine_tune_callbacks,
        verbose=1
    )
    
    return history_fine
```

```python
# Main training function
def train_model():
    """Main function to train the model"""
    
    print("📊 Creating data generators...")
    train_generator, validation_generator = create_data_generators()
    
    print(f"Training samples: {train_generator.samples}")
    print(f"Validation samples: {validation_generator.samples}")
    print(f"Classes found: {train_generator.class_indices}")
    
    print("\n🏗️ Creating model...")
    model = create_model()
    model = compile_model(model)
    
    print("\n📋 Model Summary:")
    model.summary()
    
    print(f"\n🚀 Starting training for {EPOCHS} epochs...")
    callbacks = create_callbacks()
    
    history = model.fit(
        train_generator,
        epochs=EPOCHS,
        validation_data=validation_generator,
        callbacks=callbacks,
        verbose=1
    )
    
    print("\n✅ Training completed!")
    
    # Plot training history
    plot_training_history(history)
    
    # Evaluate model
    print("\n📊 Evaluating model...")
    val_loss, val_accuracy, val_top2_acc = model.evaluate(validation_generator, verbose=0)
    print(f"Validation Accuracy: {val_accuracy:.4f}")
    print(f"Validation Top-2 Accuracy: {val_top2_acc:.4f}")
    print(f"Validation Loss: {val_loss:.4f}")
    
    # Generate predictions for confusion matrix
    validation_generator.reset()
    predictions = model.predict(validation_generator, verbose=1)
    y_pred = np.argmax(predictions, axis=1)
    y_true = validation_generator.classes
    
    # Plot confusion matrix
    plot_confusion_matrix(y_true, y_pred, CLASS_NAMES)
    
    # Print classification report
    print("\n📈 Classification Report:")
    print(classification_report(y_true, y_pred, target_names=CLASS_NAMES))
    
    # Save final model
    model.save('civic_classifier_final.h5')
    print("\n💾 Model saved as 'civic_classifier_final.h5'")
    
    return model, history
```

```python
# Load trained model function
def load_trained_model(model_path='civic_classifier_final.h5'):
    """Load a pre-trained model"""
    
    try:
        model = keras.models.load_model(model_path)
        print(f"✅ Model loaded from {model_path}")
        return model
    except:
        print(f"❌ Could not load model from {model_path}")
        return None

# Usage Examples
if __name__ == "__main__":
    print("🎯 Civic Categories Classification System")
    print("=" * 50)
    
    # Check if dataset directory exists
    if not os.path.exists(DATA_DIR):
        print(f"⚠️  Dataset directory '{DATA_DIR}' not found!")
        print("Please create the following directory structure:")
        print("""
dataset/
├── train/
│   ├── garbage_bin/
│   ├── road/
│   ├── sidewalk_path/
│   ├── drainage_system/
│   ├── parking_areas/
│   ├── street_lights/
│   └── park/
└── validation/
    ├── garbage_bin/
    ├── road/
    ├── sidewalk_path/
    ├── drainage_system/
    ├── parking_areas/
    ├── street_lights/
    └── park/
        """)
    else:
        print("📁 Dataset directory found!")
        
        # Option 1: Train new model
        print("\n🔄 Choose an option:")
        print("1. Train new model")
        print("2. Load existing model for prediction")
        print("3. Fine-tune existing model")
        
        # For demonstration, let's show how to train
        print("\n🚀 Training new model...")
        model, history = train_model()
        
        # Example: Predict on custom image
        print("\n🔍 Example: Predict custom image")
        print("To predict on your custom image, use:")
        print("""
# Load model (if not already loaded)
model = load_trained_model('civic_classifier_final.h5')

# Predict single image
image_path = 'path/to/your/image.jpg'
predicted_class, confidence, top_3 = predict_single_image(model, image_path)

print(f"Prediction: {predicted_class}")
print(f"Confidence: {confidence:.2%}")
print(f"Top 3 predictions: {top_3}")
        """)
```

```python
# Additional utility functions for batch prediction
def predict_batch_images(model, image_folder, output_csv='predictions.csv'):
    """Predict classes for all images in a folder"""
    
    import pandas as pd
    import glob
    
    results = []
    image_extensions = ['*.jpg', '*.jpeg', '*.png', '*.bmp', '*.tiff']
    
    for ext in image_extensions:
        image_paths = glob.glob(os.path.join(image_folder, ext))
        
        for image_path in image_paths:
            try:
                predicted_class, confidence, top_3 = predict_single_image(
                    model, image_path, show_image=False
                )
                
                results.append({
                    'image_path': image_path,
                    'predicted_class': predicted_class,
                    'confidence': confidence,
                    'top_1': top_3[0][0],
                    'top_1_conf': top_3[0][1],
                    'top_2': top_3[1][0],
                    'top_2_conf': top_3[1][1],
                    'top_3': top_3[2][0],
                    'top_3_conf': top_3[2][1]
                })
                
                print(f"✅ {os.path.basename(image_path)}: {predicted_class} ({confidence:.2%})")
                
            except Exception as e:
                print(f"❌ Error processing {image_path}: {e}")
    
    # Save results to CSV
    df = pd.DataFrame(results)
    df.to_csv(output_csv, index=False)
    print(f"\n💾 Results saved to {output_csv}")
    
    return df

print("\n🎉 Civic Categories Classifier is ready!")
print("Make sure to update the DATA_DIR variable to point to your dataset location.")
```
